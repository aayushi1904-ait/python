## Example:9 -basic matplotlib table 

```python
%%writefile example9.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt

st.title("matplotlib + Streamlit Demo")

# sample data
x=np.arange(1,11)
y=np.random.randint(50,100,size=10)

# line chart
plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.xlabel("Student Index")
plt.ylabel("Student Marks")
plt.title("Marks of 10 student")
st.pyplot(plt)
plt.clf() # to reduse overlaping


#barchart
st.subheader("bar chart")
plt.bar(x,y)
plt.xlabel("Student Index")
plt.ylabel("Student Marks")
plt.title("Marks of 10 student")
st.pyplot(plt)
plt.clf()


# Scater
st.subheader("scater chart")
plt.xlabel("Student Index")
plt.ylabel("Student Marks")
plt.title("Marks of 10 student")
plt.scatter(x,y,marker="*",color="g",s=60)
st.pyplot(plt)
plt.clf()

# histogram
st.subheader("histogram")
plt.xlabel("Student Index")
plt.ylabel("Student Marks")
plt.title("Marks of 10 student")
plt.hist(y,color="blue")
st.pyplot(plt)
plt.clf()
```

```python
%%writefile example11.py
import streamlit as st
import time
st.title("Status Element Demo")
st.success("This is a success message")
st.warning("This is a warning message")
st.error("This is a erroe message")
st.info("Useful information can go here")

st.write("--")
st.subheader("progress and spinner example")

if st.button("start Long Task"):
    progress=st.progress(0)
    with st.spinner("processing....."):
        for i in range(100):
            time.sleep(0.03)
            progress.progress(i+1)
        st.success("Task completed!")
```

```python
%%writefile practice.py
import streamlit as st
import pandas as pd
from datetime import date

st.set_page_config(page_title="student Marks and Feedback form",layout="centered")

st.title("Student Marks and Feedback Form")

col1,col2=st.columns(2)

with col1:
    enroll=st.text_input("Enrollment Number")
    name=st.text_input("student Name")
with col2:
    sem=st.selectbox("semester",["3","4","5","6","7","8"])
    div=st.text_input("Division","CSE-A")
exam_date=st.date_input("Exam Date",value=date.today())

st.header("2.Marks Entry")

marks_py1=st.number_input("Python-1 marks(out of 100)",min_value=0,max_value=100,value=0)
marks_fsd1=st.number_input("FSD-1 marks(out of 100)",min_value=0,max_value=100,value=0)
marks_de=st.number_input("DE marks(out of 100)",min_value=0,max_value=100,value=0)

st.header("3.Feedback")

understanding=st.slider("how well did you understand the subject? ",1,10,7)
participation=st.radio("class participation",["low","medium","High"])
comments=st.text_area("additional comments")

if st.button("submit record"):
    if not enroll or not name:
        st.error("pls fill enrollment number and student name.")
    else:
        df=pd.DataFrame({"Enrollment":[enroll],
                         "Name":[name],
                         "Sem":[sem],
                         "Division":[div],
                         "Exam Date":[exam_date],
                         "Python-1":[marks_py1],
                         "FSD-1":[marks_fsd1],
                         "DE":[marks_de],
                         "Understanding":[understanding],
                         "Participation":[participation],
                         "Comments":[comments]
                        })
        st.success("Record submitted successfully!")
        st.subheader("preview")
        st.dataframe(df)
```
