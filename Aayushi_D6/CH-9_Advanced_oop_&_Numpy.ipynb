### 2.Operator Overloding 
### Other methods
    1.add
    2.sub
    3.mul
    4.trusdiv
    5.floordiv
    6.mod
    7.pow
### methods
    1.lt(less than)
    2.gt(greater than)
    3.le(less or equal)
    4.ge(greater or equal)
    5.eq(equal to)
    6.ne(not equal to)

```python
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __str__(self):
        return "x ="+str(self.x)+" y ="+str(self.y)
    def __add__(self,other):
        x=self.x+other.x
        y=self.y+other.y
        return Point(x,y)
    
    def __sub__(self,other):
        x=self.x-other.x
        y=self.y-other.y
        return Point(x,y)
    
    def __mul__(self,other):
        x=self.x*other.x
        y=self.y*other.y
        return Point(x,y)
    
    def __truediv__(self,other):
        x=self.x/other.x
        y=self.y/other.y
        return Point(x,y)
    
    def __floordiv__(self,other):
        x=self.x//other.x
        y=self.y//other.y
        return Point(x,y)
    
    def __mod__(self,other):
        x=self.x%other.x
        y=self.y%other.y
        return Point(x,y)
    
    def __pow__(self,other):
        x=self.x**other.x
        y=self.y**other.y
        return Point(x,y)
    
    def __lt__(self,other):
        if self.x<other.x:
            if self.y<other.y:
                return True
        return False
    
    def __gt__(self,other):
        if self.x>other.x:
            if self.y>other.y:
                return True
        return False
    
    def __le__(self,other):
        if self.x<=other.x:
            if self.y<=other.y:
                return True
        return False
    
    def __ge__(self,other):
        if self.x>=other.x:
            if self.y>=other.y:
                return True
        return False
    
    def __eq__(self,other):
        if self.x==other.x:
            if self.y==other.y:
                return True
        return False
    
    def __ne__(self,other):
        if self.x!=other.x:
            if self.y!=other.y:
                return True
        return False
```

```python
p1=Point(6,31)
p2=Point(3,4)
```

```python
p3=p1+p2
p4=p1-p2
p5=p1*p2
p6=p1/p2
p7=p1//p2
p8=p1%p2
p9=p1**p2
p10=p1<p2
p11=p1>p2
p12=p1<=p2
p13=p1>=p2
p14=(p1==p2)
p15=p1!=p2
print(p3)
print(p4)
print(p5)
print(p6)
print(p7)
print(p8)
print(p9)
print(p10)
print(p11)
print(p12)
print(p13)
print(p14)
print(p15)
```

### 3.Inheritance 
    1.single inheritance
    2.multiple inheritance
    3.multilevel inheritance
    4.heirarchical inheritance
    5.hybrid inheritance

```python
## 1.single inheritance
class Parent:
    def fun1(self):
        print("Hello Parent")
        
class Child(Parent):
    def fun2(self):
        print("Hello child")
```

```python
p=Parent()
p.fun1()
```

```python
c=Child()
c.fun2()
c.fun1()
```

```python
## 2.multiple inheritance
class Parent1:
    def fun1(self):
        print("Hello parent1")
class Parent2:
    def fun2(self):
        print("Hello parent2")
class Child(Parent1,Parent2):
    def fun3(self):
        print("Hello child")
```

```python
p=Parent1()
p.fun1()
p1=Parent2()
p1.fun2()
c=Child()
c.fun3()
c.fun2()
c.fun1()
```

```python
## 3. multilevel inheritance
class Grantparent:
    def fun1(self):
        print("Hello grantparent")
class Parent(Grantparent):
    def fun2(self):
        print("Hello parent")
class Child(Parent):
    def fun3(self):
        print("Hello child")
```

```python
g=Grantparent()
g.fun1()
p=Parent()
p.fun1()
p.fun2()
c=Child()
c.fun1()
c.fun2()
c.fun3()
```

```python
## 4.hierarchical inheritance
class Parent:
    def fun1(self):
        print("Hello parent")
        
class Child1(Parent):
    def fun2(self):
        print("Hello Child1")
        
class Child2(Parent):
    def fun3(self):
        print("Hello child2")
```

```python
p=Parent()
p.fun1()

c1=Child1()
c1.fun1()
c1.fun2()

c2=Child2()
c2.fun1()
c2.fun3()
```

```python
## hybrid inheritance
class Grantparent:
    def fun1(self):
        print("Hello grantparent")
        
class Parent1(Grantparent):
    def fun2(self):
        print("Hello parent1")
        
class Parent2:
    def fun3(self):
        print("Hello parent2")
        
class Child(Parent1,Parent2):
    def fun4(self):
        print("Hello child")
```

```python
# Dimond method
class A:
    def test(self):
        print("in class A")
        
class B(A):
    def test(self):
        print("in class B")
        
class C(A):
    def test(self):
        print("in class C")
        
class D(B,C):
    def display(self):
        print("in class D")
```

### Abstract class 

```python
from abc import ABC,abstractmethod
```

```python
class Shape(ABC):
    def __init__(self,shape_name):
        self.shape_name=shape_name
        
    @abstractmethod
    def draw(self):
        pass
    
class circle(Shape):
    def __init__(self):
        super().__init__("circle")
    def draw(self):
        print("drawing a circle")
class Triangle(Shape):
    def __init__(self):
        super().__init__("Triangle")
    def draw(self):
        print("drawing a Triangle")
```

```python
c1=circle()
c1.draw()
t1=Triangle()
t1.draw()
```

```python
# pb-657
class Employee(ABC):
    def __init__(self):
        self.id=0
        self.name=""
        self.salary=0
    def get_details(self):
        self.id=int(input("id : "))
        self.name=input("name : ")
        self.salary=int(input("salary : "))
    def display(self):
        print("employee id = ",self.id)
        print("employee name = ",self.name)
        print("employee salary = ",self.salary)
    def return_salary(self):
        return self.salary
    @abstractmethod
    def emp_id(self):
        pass
    
class Perks(Employee):
    def __init__(self):
        super().__init__()
        self.da=0
        self.hra=0
        self.pf=0
    def emp_id(self):
        return self.id
    def calculate(self):
        self.da=self.salary*0.35
        self.hra=self.salary*0.17
        self.pf=self.salary*0.12
    def display_perks(self):
        print(" DA : ",self.da)
        print(" HRA : ",self.hra)
        print(" pf : ",self.pf)
        print(" total perks : ",self.da+self.hra-self.pf)   
        
class NetSalary(Perks):
    def __init__(self):
        super().__init__()
        self.net_salary=0
    def calculate_net_salary(self):
        self.net_salary=self.salary+self.da+self.hra-self.pf
    def display_net(self):
        self.display()
        self.display_perks()
        print("total salary =",self.net_salary)
```

```python
A=NetSalary()
A.get_details()
A.calculate()
A.calculate_net_salary()
A.display_net()
```
